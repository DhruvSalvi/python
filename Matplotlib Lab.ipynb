```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'city':['Mumbai','Surat','Pune','Banglore'],
    'temp':['39','34','26','32'],
    'wind_speed':['22','35','65','46']
})
df
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.xlabel('City')
plt.ylabel('Temprature')
plt.title('Weather Table')
plt.plot(df['city'],df['temp'],'g--',label='Temprature')
plt.plot(df['city'],df['wind_speed'],'c--',label='Windspeed')

plt.legend(loc='upper left')
plt.grid()
```

```python
# This below data is about "Bar" Chart in Matplotelib.
```

```python
xpos= np.arange(len(df['city']))
#plt.xticks(ypos,df['city'])
plt.barh(xpos,df['temp'])
plt.barh(xpos,df['wind_speed'])
```

```python
plt.xlabel('City')
plt.ylabel('Temprature')
plt.title('Weather')

plt.bar(df['city'],df['wind_speed'],label='Windspeed',width=0.3)
plt.bar(df['city'],df['temp'],label='Temprature',width=0.3)

plt.legend()

plt.savefig('D:\Data science\Matplotlib\.png',pad_inches=2)
```

```python
# This below data is about "Histogram" Chart in Matplotelib.
```

```python
bsw = [122,134,132,145,45,54,68,76,145,163,80,100]
bsm = [132,144,112,105,45,74,80,176,125,163,180,150]
plt.xlabel('Frequence')
plt.ylabel('Total number of patients')
plt.title('Blood Suger')

plt.hist([bsw,bsm],bins=[80,100,125,200],rwidth=0.80,color=['c','r'],label=['Women','Men'],orientation='horizontal')

plt.legend()
plt.savefig('D:\Data science\Matplotlib\histagram.png',pad_inches=2)
```
