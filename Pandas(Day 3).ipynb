```python
df1 = pd.DataFrame({
    'city':['surat','navsari','bardoli'],
    'weather':['sunny','rain','snow']
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['surat','vapi','bardoli'],
    'windspeed':['9','13','18']
})
df2
```

```python
df = pd.merge(df1,df2,on='city',how='outer',indicator=True)
df
```
